```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
from pandas import DataFrame

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset

combined_df = pd.merge(mouse_metadata, study_results, on='Mouse ID', how = 'outer')

# Display the data table for preview
combined_df
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
d_mean = round(combined_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].mean(),2)
d_med = round(combined_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].median(),2)
d_var = round(combined_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].var(),2)
d_std = round(combined_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].std(),2)
d_sem = round(combined_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].sem(),2)

# This method is the most straighforward, creating multiple series and putting them all together at the end.
summary_df = pd.DataFrame({'mean': d_mean,
                          'median': d_med,
                          'variance': d_var,
                          'std': d_std,
                           'sem': d_sem,
                          })
summary_df
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin
combined_df.sort_values(by = ['Timepoint'], ascending = True)

#drop values that have 0's
combined_df = combined_df.loc[combined_df['Timepoint']!=0,:]

#drop duplicates
combined_df = combined_df.drop_duplicates(subset='Mouse ID', keep='last')

#list of drugs
list_of_drugs = ['Capomulin', 'Ramicane','Infubinol','Ceftamin']

# Put treatments into a list for for loop (and later for plot labels)

lowerqs = []
upperqs = []
iqrs = []
lower_bounds = []
upper_bounds = []
drug_counter = []


# Calculate the IQR and quantitatively determine if there are any potential outliers. 

for drugs in list_of_drugs:
    
    quartiles = combined_df.loc[(combined_df['Drug Regimen'] == drugs), 'Tumor Volume (mm3)'].quantile([.25,0.5,0.75])
    
    lowerq = round(quartiles[0.25],2)
    upperq = round(quartiles[0.75],2)
    iqr = round(upperq - lowerq,2)
    lower_bound = round(lowerq - (1.5*iqr),2)
    upper_bound = round(upperq + (1.5*iqr),2)
    
    counter = combined_df.loc[(combined_df['Drug Regimen'] == drugs), 'Tumor Volume (mm3)'].count()
    
    lowerqs.append(lowerq)
    upperqs.append(upperq)
    iqrs.append(iqr)
    lower_bounds.append(lower_bound)
    upper_bounds.append(upper_bound)
    drug_counter.append(counter)
    
    
quartile_summary = pd.DataFrame({"List of Drugs":list_of_drugs,
                                 "Drug Count": drug_counter,
                                "lower Q": lowerqs,
                                "upper Q": upperqs,
                                "IQRs": iqrs,
                                "Lower Bound": lower_bounds,
                                "Upper Bound": upper_bounds})
quartile_summary
quartile_summary.set_index("List of Drugs")
```

```python
# Create empty list to fill with tumor vol data (for plotting)

top4_drugs = combined_df.loc[(combined_df['Drug Regimen'] == list_of_drugs[0])|
                             (combined_df['Drug Regimen'] == list_of_drugs[1])|
                             (combined_df['Drug Regimen'] == list_of_drugs[2])|
                             (combined_df['Drug Regimen'] == list_of_drugs[3])]    
top4_drugs

# Determine outliers using upper and lower bounds
for d in range(len(list_of_drugs)):
    outliers = top4_drugs.loc[(top4_drugs['Drug Regimen'] == list_of_drugs[d])]
    outliers = outliers.loc[(outliers['Tumor Volume (mm3)'] < lower_bounds[d])|(outliers['Tumor Volume (mm3)'] > upper_bounds[d])]
    
    display(outliers)

```

```python
# Locate the rows which contain mice on each drug and get the tumor volumes
top4_drugs.boxplot(column = 'Tumor Volume (mm3)', by = 'Drug Regimen', flierprops = dict(markerfacecolor = 'g', marker = 'o'))

plt.title('')
plt.xlabel('Drug Regimen')
plt.ylabel('Tumor Volume (mm3)')
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen

#Create a data set that's just Capomulin
capomulin = combined_dff.loc[(combined_dff["Drug Regimen"] == 'Capomulin')]

#get the x_axis data
tumorvolume = capomulin.groupby('Mouse ID')['Tumor Volume (mm3)'].mean()
#get the y_axis data
mouseweight = capomulin.groupby('Mouse ID')['Weight (g)'].mean()

#plot the scatterplot
plt.scatter(mouseweight, tumorvolume, marker="o", facecolor='green', edgecolors='green')

#get the slope, intercept, rvalue, pvalue, and standard error
(slope, intercept, rvalue, pvalue, stderr) = st.linregress(mouseweight, tumorvolume)
regress_values = mouseweight * slope + intercept

#line equation
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.plot(mouseweight,regress_values,"r-")

#get the correlation
correlation = round(st.pearsonr(mouseweight,tumorvolume)[0],2)
r_sq= round(correlation**2,2)

#get title, xlabel, ylabel, linear equation, correlation, and r^2
plt.xlabel("Weight (g)")
plt.ylabel("Tumor Volume (mm3)")
plt.title("Average Weight of Capomulin v Tumor Volume")
print(f'The slope is {line_eq}')
print(f'The correlatation is {correlation}')
print(f'R Squared {r_sq}')
```
